```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def f1(y1,y2):
    return -y2 + y1 * (1 - np.sqrt(y1**2 + y2**2))
    
def f2(y1,y2):
    return y1 + y2 * (1- np.sqrt(y1**2 + y2**2))

def f(t,y):
    return [f1(y[0],y[1]),f2(y[0],y[1])]
y_0 = [1,0]
a, b = 1, 6
```

```python
y = [f1,f2]
t = np.linspace(a,b,100)
sol = solve_ivp(f,(a,b),y_0,t_eval=t)
print(sol)
plt.plot(sol.t,sol.y[0])
plt.plot(sol.t,sol.y[1])
```
